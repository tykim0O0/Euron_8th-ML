### 15. What is the result of the following expression?

(실행 후, 결과에 대한 주석 작성해주세요. ex. 출력 결과에 대한 이유)

```python
0 * np.nan
np.nan == np.nan
np.inf > np.nan
np.nan - np.nan
np.nan in set([np.nan])
0.3 == 3 * 0.1
```

````python
0 * np.nan
# NaN(Not a Number)은 숫자가 아니기 때문에 다른 숫자들과 연산하면 결과값으로 nan이 반환됨
````

````python
np.nan == np.nan
# nan은 다른 수와 순서를 가질 수 없으므로(Unordered) 비교 연산하면 !=를 제외하고 False가 출력된다
````

````python
np.inf > np.nan
# 지정 데이터타입의 최솟값, 최댓값을 넘어가면 음의 무한(-np.inf), 양의 무한(np.inf)으로 표기됨
# 마찬가지로 nan는 unordered 데이터이므로 False가 출력된다
````

````python
np.nan - np.nan
# nan는 수학적으로 정의되지 않은 값이므로, nan가 다른 nan 값과 동일하다는 보장이 없다
````

````python
print(np.nan in set([np.nan]))
# in 연산자는 == 비교연산이 아닌, is 연산(객체의 ID를 비교하는 연산)을 수행하므로,
# set([np.nan])에 저장된 nan 객체와 np.nan 객체는 같은 ID를 가지고 있다.
# 따라서, 같은 nan 객체이기 때문에 True가 반환된다.
print([id(s) for s in set([np.nan])], id(np.nan))
````

````python
print(0.3 == 3 * 0.1)
# float(부동소수점) 연산은 아주 작은 오차를 발생시킬 수 있다.
print(0.3)
print(3*0.1)
````

### 16. Normalize a 5x5 random matrix.

(hint: (x - mean) / std)

#### ✅출력 예시


```
[[-1.4765971   0.9582795  -1.29910242  1.0250981  -0.79801651]
 [ 0.57945383 -1.28559078  0.02733238  0.99675052  1.2702088 ]
 [-0.08513084 -1.4686812  -0.78813593 -0.34559303 -1.15052328]
 [ 0.98183813 -1.41255344  0.37190169  1.05697419 -0.02668639]
 [-1.14466851  0.98911205  1.14499256  0.69454882  1.18478886]]
```

````python
# 위의 normalize = Z-Score 정규화: 평균이 0, 표준편차가 1인 형태로 변환하는 방법
array1=np.random.rand(5,5)
array1=(array1-np.mean(array1))/np.std(array1)
print(array1)

# 확인
print()
print(f"평균: {np.mean(array1)}, 표준편차: {np.std(array1)}")
````
